```python
import tkinter as tk
from tkinter import *
import os
from collections import Counter
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
import _pickle as c
```

```python
window = tk.Tk()
window.title("Content Based Spam Mail Detection")
def make_dict(data_dir):
    path = [os.path.join(data_dir,f) for f in os.listdir(data_dir)]
    
    words=[]
    c= len(path)
    
    for email in path:
        f= open(email,encoding="latin-1")
        blob = f.read()
        words = words+blob.split(" ")
        c-=1
    for i in range(len(words)):
        if not words[i].isalpha():
            words[i]=""
            
    dictionary = Counter(words)
    del dictionary[""]
    return (dictionary.most_common(3000))
def make_datset(dictionary,data_dir):
    path = [os.path.join(data_dir,f) for f in os.listdir(data_dir)]
    
    feature_set=[]
    labels=[]
    c= len(path)
    
    for email in path:
        data=[]
        f= open(email,encoding="latin-1")
        words = f.read().split(' ')
        for entry in dictionary:
            data.append(words.count(entry[0]))
        feature_set.append(data)
        if "ham" in email:
            labels.append(0)
        if "spam" in email:
            labels.append(1)
        #print (c)
        c -=1
    return feature_set,labels




def loa(clf_file):
    with open(clf_file,'rb') as fp:
        clf = c.load(fp)
    return clf

d= make_dict("emails")
features,labels = make_datset(d,"emails")
x_train,x_test,y_train,y_test = tts(features,labels,test_size=0.2)

clf1 = MultinomialNB()
clf1.fit(x_train,y_train)
preds = clf1.predict(x_test)
print (accuracy_score(y_test,preds))
filename="text-classifier.mdl"
c.dump(clf1,open(filename,'wb'))





def get_inputs():
    clf = loa("text-classifier.mdl")
    d= make_dict("emails")
    
    
    features=[]
        
    inp = txt.get()
        
     

    for word in d:
        features.append(inp.count(word[0]))

    res=clf.predict([features])
    print(res)
    
    if res==0:
        t2_var.set("Not Spam")
    else:
        t2_var.set("Spam")

#Label 1
l1=tk.Label(window,text="WRITE YOUR MESSAGE HERE",font=("CALIBRI",20))
#l1.grid(column=0, row=0)
l1.pack()
#Scrolled Text
txt = DoubleVar()
txt=tk.Entry(window,width=40,bd=5)
txt.pack()
#t1=tk.Entry(window,width=50,bd=5)
#t1.grid(column=0, row=1)
#t1.pack()

#Button




    
b1=tk.Button(window,text="Check",font=("CALIBRI BOLD",15),command=get_inputs)
b1.pack()

#Label 2
l2=tk.Label(window,text="RESULT",font=("CALIBRI",20))
#l2.grid(column=0, row=3)
l2.pack()
#TextField 2
t2_var = DoubleVar()
t2=tk.Entry(window,width=40,bd=5,textvariable=t2_var)
#t2.grid(column=0, row=4)
t2.pack()
window.geometry("800x200")
window.mainloop()









    



    
     

```
